```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from utils import fourier
```

```python
import os
import glob


f = glob.glob('/home/ruairi/Documents/mikex/filtering_*.mat')[0]
data = loadmat(f)
```

```python
fs = data['fs'][0][0]
x = data['x'].flatten()
y = data['y'].flatten()
t = np.arange(1, len(x) + 1) * (1/fs)

xX, xf = fourier(x, fs)
yX, yf = fourier(y, fs)

f, a =  plt.subplots(1, 2, figsize=(8, 3))
a[0].plot(t, x, label='unfilted')
a[0].plot(t, y, label='filtered')
a[1].plot(xf, np.abs(xX), label='unfilted')
a[1].plot(yf, np.abs(yX), label='filtered')
a[1].set_xlim([0, 60])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# bandpass 8 - 30
from scipy.signal import filtfilt, firls, firwin

def bandpass(bands, fs, order=5, pass_zero=False):
    order = int(np.round(order * (fs / bands[1])))
    order = order + 1 if order % 2 == 0 else order
    kernel = firwin(order, bands, fs=fs, pass_zero=pass_zero)
    kX, kf = fourier(kernel, fs)
    return kernel, kX, kf

freqs = [5, 34]
kernel_1, kX, kf = bandpass(bands=freqs, fs=fs, order=20, pass_zero=False)
f, a = plt.subplots(1, 2)
a[0].plot(kernel)
a[1].plot(kf, np.abs(kX))
a[1].set_xlim([0, 80])
```

```python
freqs = [18, 25]
kernel_2, kX, kf = bandpass(bands=freqs, fs=fs, order=20, pass_zero=True)

f, a = plt.subplots(1, 2)
a[0].plot(kernel_2)
a[1].plot(kf, np.abs(kX))
a[1].set_xlim([0, 80])
```

```python
sig2 = x
sig1 = filtfilt(kernel_1, 1, sig2)
sig = filtfilt(kernel_2, 1, sig1)
```

```python

xX, xf = fourier(x, fs)
yX, yf = fourier(y, fs)
sigX, sigf = fourier(sig, fs)

f, a =  plt.subplots(1, 2, figsize=(8, 3))
a[0].plot(t, x, label='unfilted')
a[0].plot(t, y, label='filtered')
a[0].plot(t, sig, label='my_filter')

a[1].plot(xf, np.abs(xX), label='unfilted')
a[1].plot(yf, np.abs(yX), label='filtered')
a[1].plot(sigf, np.abs(sigX), label='my_filter')
a[1].set_xlim([0, 60])
plt.legend()
plt.tight_layout()
plt.show()
```
